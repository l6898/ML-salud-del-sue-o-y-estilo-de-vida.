### Dividimos los datos en datos de entrenamiento y datos de test

```python
datos_entrenamiento = df.sample(frac = 0.8, random_state = 0)
datos_test = df.drop(datos_entrenamiento.index)
```

```python
datos_entrenamiento
```

### Separamos la variable que queremos predecir

```python
 etiquetas_entrenamiento = datos_entrenamiento.pop ('Quality of Sleep')
 etiquetas_test = datos_test.pop ('Quality of Sleep')
```

```python
datos_entrenamiento
```

```python
etiquetas_entrenamiento
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(datos_entrenamiento, etiquetas_entrenamiento)
```

```python
predicciones = modelo.predict(datos_test)
predicciones
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(etiquetas_test,predicciones))
print ("Error porcentual : %f" % (error*100))
```
